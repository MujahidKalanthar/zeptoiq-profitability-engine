```python
import pandas as pd
import numpy as np
import plotly.express as px

# Load dataset
df = pd.read_csv("../data/synthetic/orders_data.csv")

df.head()
```

```python
# Base metrics

base_orders = len(df)

base_profit = df["contribution_margin"].sum()

# Simulate demand spike

demand_multiplier = 1.25

projected_orders = (
    base_orders * demand_multiplier
)

projected_profit = (
    base_profit * 1.15
)

print("Base Orders:", base_orders)

print("Projected Orders:", round(projected_orders))

print("Base Profit:", round(base_profit,2))

print("Projected Profit:",
      round(projected_profit,2))
```

```python
scenario_df = pd.DataFrame({

    "Scenario": [
        "Current",
        "Demand Spike"
    ],

    "Profit": [
        base_profit,
        projected_profit
    ]
})

fig = px.bar(

    scenario_df,

    x="Scenario",

    y="Profit",

    color="Scenario",

    title="Demand Spike Profitability Impact"

)

fig.show()
```

```python
# Increased discount pressure

current_discount = df["discount"].mean()

increased_discount = (
    current_discount * 1.5
)

discount_impact_profit = (

    df["gross_margin"].sum()

    + df["delivery_fee"].sum()

    - df["rider_cost"].sum()

    - df["dark_store_cost"].sum()

    - (increased_discount * len(df))
)

print("Current Avg Discount:",
      round(current_discount,2))

print("Aggressive Discount Scenario:",
      round(increased_discount,2))

print("Projected Profit Under Discount War:",
      round(discount_impact_profit,2))
```

```python
discount_df = pd.DataFrame({

    "Scenario": [
        "Current Pricing",
        "Discount War"
    ],

    "Profit": [
        base_profit,
        discount_impact_profit
    ]
})

fig = px.bar(

    discount_df,

    x="Scenario",

    y="Profit",

    color="Scenario",

    title="Discount War Profitability Impact"

)

fig.show()
```

```python
# Simulate labor regulation impact

higher_rider_cost = (
    df["rider_cost"] * 1.20
)

labor_impact_profit = (

    df["gross_margin"].sum()

    + df["delivery_fee"].sum()

    - higher_rider_cost.sum()

    - df["dark_store_cost"].sum()

    - df["discount"].sum()
)

print("Current Rider Cost:",
      round(df["rider_cost"].mean(),2))

print("Regulated Rider Cost:",
      round(higher_rider_cost.mean(),2))

print("Projected Profit Under Regulation:",
      round(labor_impact_profit,2))
```

```python
labor_df = pd.DataFrame({

    "Scenario": [
        "Current Operations",
        "Labor Regulation"
    ],

    "Profit": [
        base_profit,
        labor_impact_profit
    ]
})

fig = px.bar(

    labor_df,

    x="Scenario",

    y="Profit",

    color="Scenario",

    title="Labor Regulation Profitability Impact"

)

fig.show()
```

```python
# Simulate retention decline

retention_impact_multiplier = 0.90

delayed_delivery_profit = (
    base_profit * retention_impact_multiplier
)

delay_df = pd.DataFrame({

    "Scenario": [
        "Current SLA",
        "Delivery Delays"
    ],

    "Profit": [
        base_profit,
        delayed_delivery_profit
    ]
})

fig = px.bar(

    delay_df,

    x="Scenario",

    y="Profit",

    color="Scenario",

    title="Delivery Delay Impact on Profitability"

)

fig.show()
```

```python
master_scenarios = pd.DataFrame({

    "Scenario": [

        "Current Operations",
        "Demand Spike",
        "Discount War",
        "Labor Regulation",
        "Delivery Delays"
    ],

    "Projected Profit": [

        base_profit,
        projected_profit,
        discount_impact_profit,
        labor_impact_profit,
        delayed_delivery_profit
    ]
})

master_scenarios
```

```python
fig = px.bar(

    master_scenarios,

    x="Scenario",

    y="Projected Profit",

    color="Projected Profit",

    title="Strategic Scenario Profitability Stress Test"

)

fig.show()
```
