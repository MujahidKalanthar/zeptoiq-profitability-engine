```python
import sys
sys.path.append("../src")

from pricing_engine import (
    calculate_z_score,
    recommend_delivery_fee,
    estimate_profitability
)

print("Pricing engine loaded successfully 😤")
```

```python
# Example operational inputs

demand_intensity = 8
delivery_distance = 4
customer_priority = 9
inventory_pressure = 5
basket_value = 6

# Calculate z-score

z_score = calculate_z_score(
    demand_intensity,
    delivery_distance,
    customer_priority,
    inventory_pressure,
    basket_value
)

# Recommended pricing

recommended_fee = recommend_delivery_fee(z_score)

# Estimate profitability

profitability = estimate_profitability(
    order_value=550,
    delivery_fee=recommended_fee,
    rider_cost=45,
    dark_store_cost=30,
    discount=20
)

print("Z-Score:", z_score)

print("Recommended Delivery Fee:", recommended_fee)

print("Estimated Contribution Margin:", profitability)
```

```python
import pandas as pd

scenarios = []

for demand in range(1, 11):

    z = calculate_z_score(
        demand_intensity=demand,
        delivery_distance=4,
        customer_priority=7,
        inventory_pressure=5,
        basket_value=6
    )

    fee = recommend_delivery_fee(z)

    profit = estimate_profitability(
        order_value=500,
        delivery_fee=fee,
        rider_cost=45,
        dark_store_cost=30,
        discount=20
    )

    scenarios.append([
        demand,
        z,
        fee,
        profit
    ])

pricing_df = pd.DataFrame(
    scenarios,
    columns=[
        "Demand Intensity",
        "Z-Score",
        "Recommended Fee",
        "Estimated Profit"
    ]
)

pricing_df
```

```python
import plotly.express as px

fig = px.line(
    pricing_df,
    x="Demand Intensity",
    y="Estimated Profit",
    title="Demand Intensity vs Estimated Profitability",
    markers=True
)

fig.show()
```

```python
fig = px.bar(
    pricing_df,
    x="Demand Intensity",
    y="Recommended Fee",
    color="Recommended Fee",
    title="Dynamic Delivery Fee Recommendations"
)

fig.show()
```
