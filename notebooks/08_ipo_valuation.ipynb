```python
import pandas as pd
import numpy as np
import plotly.express as px

# Load dataset
df = pd.read_csv("../data/synthetic/orders_data.csv")

df.head()
```

```python
# Core business metrics

total_orders = len(df)

average_order_value = df["order_value"].mean()

gross_merchandise_value = (
    df["order_value"].sum()
)

average_margin = (
    df["contribution_margin"].mean()
)

total_profit = (
    df["contribution_margin"].sum()
)

print("Total Orders:", total_orders)

print("Average Order Value:", round(average_order_value,2))

print("Gross Merchandise Value (GMV):",
      round(gross_merchandise_value,2))

print("Average Contribution Margin:",
      round(average_margin,2))

print("Total Contribution Profit:",
      round(total_profit,2))
```

```python
# Valuation multiple assumptions

valuation_multiples = [1.5, 2, 2.5, 3]

valuation_data = []

for multiple in valuation_multiples:

    estimated_valuation = (
        gross_merchandise_value * multiple
    )

    valuation_data.append([
        multiple,
        estimated_valuation
    ])

valuation_df = pd.DataFrame(

    valuation_data,

    columns=[
        "GMV Multiple",
        "Estimated Valuation"
    ]
)

valuation_df
```

```python
fig = px.bar(

    valuation_df,

    x="GMV Multiple",

    y="Estimated Valuation",

    color="Estimated Valuation",

    title="GMV Multiple Valuation Scenarios"

)

fig.show()
```

```python
# IPO readiness indicators

positive_margin_orders = (

    df[df["contribution_margin"] > 0]

    ["order_id"]

    .count()
)

margin_stability = (
    positive_margin_orders / total_orders
) * 100

average_delivery_time = (
    df["delivery_time"].mean()
)

readiness_metrics = {

    "Margin Stability %":
    round(margin_stability,2),

    "Average Delivery Time":
    round(average_delivery_time,2),

    "Average AOV":
    round(average_order_value,2),

    "Average Contribution Margin":
    round(average_margin,2)
}

readiness_metrics
```

```python
# IPO readiness scoring model

score = 0

# Margin stability
if margin_stability > 70:
    score += 30
elif margin_stability > 50:
    score += 20
else:
    score += 10

# Average AOV
if average_order_value > 500:
    score += 25
elif average_order_value > 400:
    score += 15
else:
    score += 10

# Contribution margin
if average_margin > 20:
    score += 25
elif average_margin > 10:
    score += 15
else:
    score += 10

# Delivery efficiency
if average_delivery_time < 20:
    score += 20
else:
    score += 10

print("IPO Readiness Score:", score, "/ 100")
```

```python
readiness_df = pd.DataFrame({

    "Metric": [
        "Margin Stability",
        "AOV Strength",
        "Contribution Margin",
        "Delivery Efficiency"
    ],

    "Score": [
        30 if margin_stability > 70 else 20,

        25 if average_order_value > 500 else 15,

        25 if average_margin > 20 else 15,

        20 if average_delivery_time < 20 else 10
    ]
})

fig = px.bar(

    readiness_df,

    x="Metric",

    y="Score",

    color="Score",

    title="IPO Readiness Scoring Breakdown"

)

fig.show()
```

```python
sensitivity_data = []

for aov_growth in [0.9, 1.0, 1.1, 1.2]:

    projected_gmv = (
        gross_merchandise_value * aov_growth
    )

    projected_valuation = (
        projected_gmv * 2.5
    )

    sensitivity_data.append([

        aov_growth,

        projected_gmv,

        projected_valuation
    ])

sensitivity_df = pd.DataFrame(

    sensitivity_data,

    columns=[
        "AOV Growth Multiplier",
        "Projected GMV",
        "Projected Valuation"
    ]
)

sensitivity_df
```

```python
fig = px.line(

    sensitivity_df,

    x="AOV Growth Multiplier",

    y="Projected Valuation",

    markers=True,

    title="Valuation Sensitivity to AOV Growth"

)

fig.show()
```
