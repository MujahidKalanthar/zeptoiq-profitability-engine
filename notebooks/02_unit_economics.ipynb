```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px

# Load dataset
df = pd.read_csv("../data/synthetic/orders_data.csv")

# Preview data
df.head()
```

```python
# Core KPIs

total_orders = len(df)

avg_order_value = df["order_value"].mean()

avg_contribution_margin = df["contribution_margin"].mean()

avg_delivery_cost = df["rider_cost"].mean()

total_profit = df["contribution_margin"].sum()

print("Total Orders:", total_orders)

print("Average Order Value:", round(avg_order_value, 2))

print("Average Contribution Margin:", round(avg_contribution_margin, 2))

print("Average Rider Cost:", round(avg_delivery_cost, 2))

print("Total Contribution Profit:", round(total_profit, 2))
```

```python
city_profit = (
    df.groupby("city")["contribution_margin"]
    .mean()
    .reset_index()
)

fig = px.bar(
    city_profit,
    x="city",
    y="contribution_margin",
    title="Average Contribution Margin by City",
    color="contribution_margin"
)

fig.show()
```

```python
fig = px.scatter(
    df,
    x="order_value",
    y="contribution_margin",
    color="customer_type",
    title="Order Value vs Contribution Margin",
    opacity=0.6
)

fig.show()
```
