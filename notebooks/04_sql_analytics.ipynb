```python
import pandas as pd
import sqlite3

# Load CSV
df = pd.read_csv("../data/synthetic/orders_data.csv")

# Create SQLite database
conn = sqlite3.connect("../data/processed/zepto.db")

# Store table
df.to_sql(
    "orders",
    conn,
    if_exists="replace",
    index=False
)

print("Database created successfully 😤")
```

```python
query = """

SELECT
    city,
    ROUND(AVG(contribution_margin),2) AS avg_margin,
    ROUND(AVG(order_value),2) AS avg_order_value,
    COUNT(*) AS total_orders

FROM orders

GROUP BY city

ORDER BY avg_margin DESC

"""

city_analysis = pd.read_sql_query(query, conn)

city_analysis
```

```python
import plotly.express as px

fig = px.bar(
    city_analysis,
    x="city",
    y="avg_margin",
    color="avg_margin",
    title="Average Contribution Margin by City"
)

fig.show()
```

```python
query = """

SELECT
    customer_type,
    ROUND(AVG(contribution_margin),2) AS avg_profit,
    ROUND(AVG(order_value),2) AS avg_aov,
    COUNT(*) AS total_orders

FROM orders

GROUP BY customer_type

ORDER BY avg_profit DESC

"""

customer_analysis = pd.read_sql_query(query, conn)

customer_analysis
```

```python
fig = px.scatter(
    customer_analysis,
    x="avg_aov",
    y="avg_profit",
    size="total_orders",
    color="customer_type",
    title="Customer Segment Profitability"
)

fig.show()
```
